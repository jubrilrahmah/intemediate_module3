```python
#  counting the no of unique diagnosis in the dataset

# df_c.groupby("diagnosis").count() #counting using the groupby method
df_d["diagnosis"].value_counts() #

```

based on the dataset, there are two different main diagnosis type, whixh is the B and the M type

```python
#using piechat to describe the ratio of B to M diagnosis
y= df_d[["diagnosis"]].groupby("diagnosis") # 
x= df_d["diagnosis"].value_counts()
plt.pie(x , labels= y, shadow= True, autopct="%1.2f%%")

plt.title("Count of unique diagnosis")

```

using a pie to represent the unique diagnosis type in the data set. from the piechart, it can be proof that the B diagnosis type is more than the M type
